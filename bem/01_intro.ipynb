# Boundary Element Methods

Boundary element methods reformulate boundary value problems as integral equations on the boundary. The unknown is a boundary density, and the solution in the domain is recovered from a layer potential.

## Advantages and Challenges

| Advantages | Challenges |
| --- | --- |
| only the boundary is meshed | system matrices are dense |
| exterior domains need no artificial boundary | singular integrals need special quadrature |
| the solution can be evaluated throughout the domain | a fundamental solution must be known |

## From PDE to Boundary Integral Equation

Let $\Omega$ be the domain and $\Gamma=\partial\Omega$ its boundary. Consider the Helmholtz problem

$$
  \Delta u + \kappa^2 u = 0 \quad \text{in } \Omega,
  \qquad
  u = g \quad \text{on } \Gamma,
$$

For a single layer representation, the solution is

$$
  u(x) = (\widetilde V\rho)(x)
  = \int_\Gamma G_\kappa(x,y)\,\rho(y)\,d\sigma_y,
  \qquad
  x \in \Omega, \qquad G_\kappa(x,y) = \frac{e^{i\kappa |x-y|}}{4 \pi |x-y|}.
$$

Here $y$ is a source point on $\Gamma$, $x$ is the evaluation point, $G_\kappa$ is the fundamental solution, and $\rho$ is the unknown boundary density. The symbol $\widetilde V$ denotes the potential in the domain.

Taking its trace on $\Gamma$ gives the boundary operator $V$ and the equation for the density:

$$
  V\rho = g \quad \text{on } \Gamma.
$$

## Boundary Discretization

Choose a finite dimensional boundary space and approximate the density by

$$
X_h = \operatorname{span}\{\varphi_1,\dots,\varphi_N\},
\qquad
\rho_h = \sum_{j=1}^N c_j\varphi_j.
$$

The Galerkin variational formulation is: find $\rho_h \in X_h$ such that

$$
  \langle V\rho_h, \eta_h \rangle_\Gamma
  = \langle g, \eta_h \rangle_\Gamma
  \qquad \forall\,\eta_h \in X_h,
$$

Choosing $\eta_h=\varphi_i$ gives the linear system

$$
\sum_{j=1}^N A_{ij}c_j=b_i,
\qquad A\mathbf c=\mathbf b,
$$

where

$$
A_{ij}=\int_\Gamma\int_\Gamma
\varphi_i(x)G_\kappa(x,y)\varphi_j(y)
\,d\sigma_y\,d\sigma_x,
\qquad
b_i=\int_\Gamma g(x)\varphi_i(x)\,d\sigma_x.
$$

## Why Fast Algorithms Are Needed

Every boundary basis function interacts with every other one through $G_\kappa(x,y)$. Therefore, $A$ is dense:

- storing the matrix costs $O(N^2)$,
- a direct matrix vector product costs $O(N^2)$.

The fast multipole method keeps nearby interactions direct and groups distant interactions through expansions. This avoids forming the full dense matrix and accelerates its application.
